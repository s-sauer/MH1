### Plots zu Aufgabe 3 (Reaktion 1. und 2. Ordnun) auf dem gleichen Übungsblatt:

```python
A0 = .1 # mol/L
t_half = 1 # halflife in s
beta = np.log(2)/t_half
k = 1/(A0*t_half)

t = np.linspace(0,10*t_half,1001)

A0_1 = A0 * np.exp(-beta*t)
A0_2 = A0 / (A0 * k * t + 1)

plt.plot(t, A0_1, t, A0_2)
plt.xlabel('Zeit [s]')
plt.ylabel('Konzentration [mol/L]')
plt.legend(['1. Ordnung', '2. Ordnung'])
plt.show()
```

### Plots zu Aufgabe 4 (Torricelli) auf dem gleichen Übungsblatt:

```python
a = 2 # 
V0 = 9 # Initial volume

t = np.linspace(0,6,1001)

def V(t):
    return(np.sqrt(V0) - a/2*t)**2

V_t_A = V(t)
V_t_A[t>3] = np.nan

V_t_B = V(t)
V_t_B[t<=3] = np.nan

plt.plot(t, V_t_A, '-b', t, V_t_B, ':b')
plt.xlabel('t')
plt.ylabel('V(t)')
```
