# Simulation of squared sum score $S^2$ of $n$ dice

### 1. Import libraries

```r
# nothing to import for R
```

### 2. Set $n$ (no. of dice) and $M$ (no. of samples)

```r
n = 12 # no of dice
M = 1e5 # no of samples ("Monte-Carlo runs")
```

### 3. Generate $M$ random samples of $n$ dice

```r
dat_as_list = sample(1:6, n*M, replace = TRUE)
dat = matrix(dat_as_list, nrow = M)  # convert to table (='matrix')
dat[0:5,]  # show row 1 to 5
```

### 4. Calculate sum score $S$ and its square $S^2$ for each sample

```r
S = rowSums(dat)
S2 = S*S
```

### 5. Plot distribution

```r
hist(S, density=TRUE, main='Histogram of sum score S')
```

```r
hist(S2, density=TRUE, main='Histogram of squared sum score S^2')
```

### 6. Calculate mean values $\bar{S}$ and $\bar{S^2}$ and compare to expected values $\mathbb{E}(S)$ and $\mathbb{E}(S^2)$

```r
mean_S  = mean(S)
mean_S2 = mean(S2)
E_S  = n*7/2
E_S2 = n*91/6 + n*(n-1)*49/4
Var_S = E_S2 - E_S**2
sprintf('Mean of S   = %.3f; cp. with expected value = %.3f', mean_S, E_S)
sprintf('Mean of S^2 = %.3f; cp. with expected value = %.3f', mean_S2,E_S2)
```
