```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

## Schwingungsgleichung

$$  \ddot x(t) + 2 \delta \dot x(t) + \omega_0^2 x(t) = 0$$

mit $ \omega_0 = \frac{2\pi}{T} $

```python
# parameter values and inital values
delta = 0.1 # unit: 1/s
omega0 = 0.5 # unit: 1/s
x0 = 1 # m
v0 = 0 # m/s
T_max = 30 # max. simulation time in s
```

```python
# define system of ordinary differential equations (ODE)
def rhs(t, y):
    [x, v] = y
    dxdt = v
    dvdt = - 2 * delta * v - omega0**2 * x

    return [dxdt, dvdt]
```

```python
solution = solve_ivp(rhs, [0,T_max], [x0, v0], t_eval=np.linspace(0,T_max,1001))
```

```python
plt.plot(solution.t, solution.y.T)
plt.legend(['x','v'])
plt.xlabel('Zeit [s]')
plt.grid('on')
plt.ylabel('Auslenkung in m / Geschwindigkeit in m/s')
```
