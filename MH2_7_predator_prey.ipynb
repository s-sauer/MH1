```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

## Preditor-prey model (a.k.a. Lotka-Volterra model)

https://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equations

\begin{align}
 \frac{dx}{dt} &= a \ x - b \ x \cdot y, \\
 \frac{dy}{dt} &= d \ x \cdot y - c \ y,
\end{align}

* $x$ ... prey (e.g., rabbits)
* $y$ ... preditor (e.g., foxes)

```python
# parameter values and inital values
a = 1 # reproduction rate of rabbits, unit: 1/year
b = 1 # rabbit reduction rate, unit: 1/(fox*year)
c = 1 # fox reduction rate, unit: 1/year
d = 1 # fox reproduction rate, unit: 1/(rabbit*year)
rabbits_init = 2 #
foxes_init = 1 #
T_max = 30 # max. simulation time in years
```

```python
# define system of ordinary differential equations (ODE)
def rhs(t, y):
    [rabbits, foxes] = y
    dRdt = +a*rabbits - b*foxes*rabbits
    dFdt = -c*foxes + d*foxes*rabbits
    return [dRdt, dFdt]
```

```python
solution = solve_ivp(rhs, [0,T_max], [rabbits_init, foxes_init], t_eval=np.linspace(0,T_max,1001))
```

```python
plt.plot(solution.t, solution.y.T)
plt.legend(['rabbits','foxes'])
plt.xlabel('Time [years]')
plt.ylabel('Number')
```
